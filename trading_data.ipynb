```python
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get data stock data
btc = yf.download("BTC-USD", start="2020-1-1", end="2024-1-1", interval="1d")

# Add average candle price
btc['avg'] = np.minimum(btc['Open'], btc['Close']) + np.abs(btc['Open'] - btc['Close'])/2

# Store average candle prices
avg_prices = np.round(list(btc['avg']), 2)


# Store dates
dates = list(btc.index)
dates = [time.date() for time in dates]

btc
```

```python
plt.figure(figsize=(10,5))
plt.plot(avg_prices)
plt.grid('on')
plt.xlabel('Date 2020-2023')
plt.ylabel('BTC Price (USD)')
```
