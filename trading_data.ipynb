```python
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# Get data stock data
aapl = yf.download("AAPL", start="2023-1-1", end="2024-1-1", interval="1d")

# Add average candle price
aapl['avg'] = np.min(aapl['Open']) + np.abs(aapl['Open'] - aapl['Close'])/2

# Store average candle prices
avg_prices = np.round(list(aapl['avg']), 2)

aapl
```
